```python
import pandas as pd
```

```python
routes = pd.read_csv('./data/routes.txt')
stops = pd.read_csv('./data/stops.txt')
trips = pd.read_csv('./data/trips.txt')
stop_times = pd.read_csv('./data/stop_times.txt')

stop_lookup = stops.set_index('stop_id')[['stop_name', 'stop_lat', 'stop_lon']].to_dict('index')
```

```python
stop_lookup
```

```python
trips.groupby('route_id')['trip_id'].first().to_dict()
```

```python
# Build a dictionary: route_id → example trip_id
route_to_trip = trips.groupby('route_id')['trip_id'].first().to_dict()

print(f'Found {len(route_to_trip)} routes.')
```

```python
import folium

# Select a route_id to plot:
selected_route_id = list(route_to_trip.keys())[0]  

print(f"Plotting route_id: {selected_route_id}")
example_trip_id = route_to_trip[selected_route_id]

# Get ordered stop sequence
trip_stops = stop_times[stop_times['trip_id'] == example_trip_id].sort_values('stop_sequence')

# Build list of coordinates
trip_coords = []
for stop_id in trip_stops['stop_id']:
    if stop_id in stop_lookup:
        lat = stop_lookup[stop_id]['stop_lat']
        lon = stop_lookup[stop_id]['stop_lon']
        trip_coords.append((lat, lon))

# Plot
if trip_coords:
    m = folium.Map(location=trip_coords[0], zoom_start=6)

    folium.PolyLine(trip_coords, color='blue', weight=4, opacity=0.8).add_to(m)

    for (lat, lon), stop_id in zip(trip_coords, trip_stops['stop_id']):
        stop_name = stop_lookup[stop_id]['stop_name']
        folium.Marker(location=(lat, lon), popup=stop_name).add_to(m)

    m
else:
    print("No coordinates found for this trip.")
```

```python
bercy_stop = stops[stops['stop_name'].str.contains("Bercy", case=False, na=False)]
bercy_stop_id = bercy_stop['stop_id'].values[0]

print(f"Paris (Bercy) stop_id = {bercy_stop_id}")
```

```python
# Group stop_times by trip_id, and get last stop_id for each trip
last_stops = stop_times.sort_values('stop_sequence').groupby('trip_id').last().reset_index()

# Filter trips where last stop_id == Paris (Bercy) stop_id
trips_ending_in_bercy = last_stops[last_stops['stop_id'] == bercy_stop_id]['trip_id'].tolist()

print(f"Found {len(trips_ending_in_bercy)} trips ending in Paris (Bercy)")
```

```python
# Initialize map centered on Paris
m = folium.Map(location=[48.835318, 2.380519], zoom_start=5)

# Loop over trips
for trip_id in trips_ending_in_bercy:
    trip_stops = stop_times[stop_times['trip_id'] == trip_id].sort_values('stop_sequence')
    trip_stops_with_names = trip_stops.merge(stops[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']], on='stop_id', how='left')
    
    trip_coords = list(zip(trip_stops_with_names['stop_lat'], trip_stops_with_names['stop_lon']))
    
    if trip_coords:
        folium.PolyLine(trip_coords, color='blue', weight=2, opacity=0.5).add_to(m)
```
